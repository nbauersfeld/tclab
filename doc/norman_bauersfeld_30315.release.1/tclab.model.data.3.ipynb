#### 1st option of data build 

N cycles and a 360s period of time  
switch with 20s delay, followed by 100s up and 240s down  

```python
%%time

data = _tclab._4_data(cycles=20)
```

#### save model

```python
name = "model.4.7"

fname = os.path.join(os.getcwd(),"data","tclab.%s.csv"%(name))
data.to_csv(fname,index=False)
```

```python
import numpy as np
from _tclab import _tclabutils

# time
t = data["Time(ns)"].to_numpy().astype(np.int64)
t = ((t-t[0])/1e9).astype(np.int64)
# input
u = data["Q1"].to_numpy()
# output
y = data["T1"].to_numpy()

# plot 
#  input of u (switch)
#  output of y (temperature)

_tclabutils.plot(t,u,y,title="%s"%(name))
```

#### 2nd option of data build  

random generated switching states with a variable height  

```python
from _helpers import Helpers_

Q1 = Helpers_.rectangle_impulse_(60*20, 50, heating=10, cooling=40)

# -----------------------------

import matplotlib.pyplot as plt

plt.rcParams["font.size"] = 8
plt.rcParams["figure.dpi"] = 80
plt.rcParams["lines.linewidth"] = .8

ncols,nrows = 1,1
nwidth,nheight = 16,4

fig, ax = plt.subplots(nrows=nrows,ncols=ncols,figsize=[nwidth*ncols,nheight*nrows],dpi=plt.rcParams["figure.dpi"])
fig.subplots_adjust(right=0.75)

ax.plot(Q1*100)
ax.grid(color="silver",zorder=-1,linestyle=":")
ax.set_xlabel("time(s)")
ax.set_ylabel("Q(%)")

plt.show()
plt.close()
```

#### proceed

```python
%%time

from _tclab import _tclab

data1 = _tclab._q_data(q=Q1)
```

#### save and plot

```python
import os 

name = "model.q.3"

fname = os.path.join(os.getcwd(),"data","tclab.%s.csv"%(name))
data1.to_csv(fname,index=False)

data1.head()
```

```python
import numpy as np
from _tclab import _tclabutils

# time
t = data1["Time(ns)"].to_numpy().astype(np.int64)
t = ((t-t[0])/1e9).astype(np.int64)
# input
u = data1["Q1"].to_numpy()
# output
y = data1["T1"].to_numpy()

# plot 
#  input of u (switch)
#  output of y (temperature)

_tclabutils.plot(t,u,y,title="%s"%(name))
```
